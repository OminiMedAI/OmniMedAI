## 4ï¸âƒ£ One-Click Segmentation Demo

```python
# ğŸ¯ One-click segmentation demo
print("ğŸš€ Running one-click segmentation demo...")

# Demo with first image
test_image = demo_data['images'][0]
print(f"ğŸ“ Processing: {test_image}")

# Simulate segmentation result
demo_seg_result = {
    'image_path': test_image,
    'model_used': 'auto' if len(demo_data['images']) > 2 else '3D',
    'processing_time': np.random.uniform(20, 60),
    'output_path': f"demo_output/{Path(test_image).stem}_roi.nii.gz",
    'confidence': np.random.uniform(0.8, 0.95)
}

print(f"âœ… Segmentation completed!")
print(f"ğŸ¯ Model used: {demo_seg_result['model_used']}")
print(f"â±ï¸  Processing time: {demo_seg_result['processing_time']:.1f}s")
print(f"ğŸ“ Output: {demo_seg_result['output_path']}")
print(f"ğŸ–ï¸  Confidence: {demo_seg_result['confidence']:.2f}")

# Quick visualization
plt.figure(figsize=(8, 3))
times = [demo_seg_result['processing_time']]
plt.bar(['Demo Segmentation'], times, color='skyblue', alpha=0.7)
plt.ylabel('Processing Time (seconds)')
plt.title('Quick Segmentation Demo')
plt.grid(True, alpha=0.3)
plt.show()
```

## 5ï¸âƒ£ One-Click Feature Extraction

```python
# ğŸ§¬ One-click radiomics feature extraction
print("ğŸ§¬ Running one-click radiomics extraction...")

# Demo radiomics features
demo_radiomics = {
    'Patient_001': {
        'firstorder_mean': 124.5,
        'firstorder_std': 32.1,
        'shape_volume': 15420,
        'glcm_contrast': 0.45,
        'glcm_homogeneity': 0.82,
        'texture_entropy': 3.21
    },
    'Patient_002': {
        'firstorder_mean': 98.7,
        'firstorder_std': 28.4,
        'shape_volume': 12890,
        'glcm_contrast': 0.38,
        'glcm_homogeneity': 0.87,
        'texture_entropy': 2.95
    }
}

print("âœ… Radiomics extraction completed!")
print(f"ğŸ“Š Features extracted for {len(demo_radiomics)} patients")
print(f"ğŸ“‹ Features per patient: {len(list(demo_radiomics.values())[0])}")

# Convert to DataFrame for nice display
rad_df = pd.DataFrame.from_dict(demo_radiomics, orient='index')
rad_df.index.name = 'PatientID'
rad_df.reset_index(inplace=True)

print("\nğŸ‘€ Sample radiomics features:")
display(rad_df)

# Quick feature visualization
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
fig.suptitle('Radiomics Features Overview', fontsize=14)

# Feature 1: Mean intensity
axes[0].bar(rad_df['PatientID'], rad_df['firstorder_mean'], color='lightgreen', alpha=0.7)
axes[0].set_title('Mean Intensity')
axes[0].set_ylabel('Hounsfield Units')

# Feature 2: Shape volume
axes[1].bar(rad_df['PatientID'], rad_df['shape_volume'], color='lightcoral', alpha=0.7)
axes[1].set_title('Shape Volume')
axes[1].set_ylabel('Voxels')

# Feature 3: Texture
axes[2].bar(rad_df['PatientID'], rad_df['glcm_contrast'], color='gold', alpha=0.7)
axes[2].set_title('GLCM Contrast')
axes[2].set_ylabel('Texture Value')

plt.tight_layout()
plt.show()
```

## 6ï¸âƒ£ Quick Pathology Analysis
